**B. ML modelleri üretimde nasıl dağıtılır?**

Bunun iki kısmı var:

I. İnsanların sisteme katılımı nasıl olacak, yani otomasyon derecesi ne olacak?

II. Dağıtım kurulumunda performans nasıl sağlanır? Mevcut insan/sistemleri aynı anda mı yoksa aşamalar halinde mi dağıtıp değiştirmeliyiz - yani dağıtım geçiş planları mı?

**Bir ML dağıtımının tamamen manuel sistemden tam otomasyona kadar uzanan bir yelpazesi var...**

a-  İnsan Liderliğindeki Sistem (Human-Led System)

b-  İnsan Liderliğindeki - Makine Destekli Sistem (Human-Led, Machine-Assisted System)

c- Makine Liderliğindeki - İnsan Destekli Sistem (Machine-Led, Human-Assisted System)

d- Makine Liderliğindeki - İnsan Tarafından Yönetilen Sistem (Machine-Led, Human-Governed System)

e- Makine Kontrollü Sistem (Machine-Controlled System)
